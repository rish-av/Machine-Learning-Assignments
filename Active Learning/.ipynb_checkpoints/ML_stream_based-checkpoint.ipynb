```python
#please install the following pip packages to run this code

#pip install modAL
#pip install scikit-learn
#pip insta

#Committee & uncerstainity sampling
from modAL.models import ActiveLearner
from modAL.models import Committee

#the sampling strategies
from modAL.uncertainty import classifier_uncertainty, classifier_margin, classifier_entropy
from modAL.uncertainty import entropy_sampling, uncertainty_sampling, margin_sampling


from modAL.disagreement import vote_entropy_sampling, max_disagreement_sampling
from modAL.disagreement import vote_entropy, KL_max_disagreement

#the algorithm for active learning
from sklearn.svm import SVC

import numpy as np
import pandas as pd
```

## Data Loading

```python
data_train = pd.read_csv('Data/pendigits.tra').to_numpy()
data_test = pd.read_csv('Data/pendigits.tes').to_numpy()

#data_for_training
x = data_train[:,:16]
y = data_train[:,16]

#data_for_testing
x_tes = data_test[:,:16]
y_tes = data_test[:,16]
```

## Stream Based Sampling with Uncertainity

```python
def train_least_confidence(x_train,y_train,x_tes,y_tes):
    #to ensure uniformity of the stream
    np.random.seed(459)
    
    accuracy_list = []
    num_queried = 0
    num_ignored = 0
    
    #initial 10 percent points
    idx_10_percent = np.random.choice(range(x_train.shape[0]),size=int(0.1*x_train.shape[0]),replace=False)
    x_10 = x_train[idx_10_percent]
    y_10 = y_train[idx_10_percent]
    
    x_train = np.delete(x_train,idx_10_percent,0)
    y_train = np.delete(y_train,idx_10_percent,0)
    
    #initial_learning
    learner = ActiveLearner(estimator=SVC(probability=True),
                                X_training=x_10,y_training=y_10)
    
    while learner.score(x_tes, y_tes) < 0.97:
        #choose a point randomly to check - simulate a stream
        point = np.random.choice(range(len(x_train))) 
        
        if classifier_uncertainty(learner, x_train[point].reshape(1, -1)) >= 0.11:
        #learning again after every query; might avoid it if the trainnig is time consuming
            learner.teach(x_train[point].reshape(1, -1), y_train[point].reshape(-1, ))
            updated_acc = learner.score(x_tes, y_tes)
            accuracy_list.append(updated_acc)
            num_queried += 1
            print(f"Point updated: {point}; Accuracy achieved: {updated_acc}") 
        else:
            num_ignored += 1
            print(f"Ignored this point from labelling: {point}")
    return accuracy_list, num_queried, num_ignored
```

```python
def train_margin_sampling(x_train,y_train,x_tes,y_tes):
    #to ensure uniformity of the stream
    np.random.seed(459)
    
    accuracy_list = []
    num_queried = 0
    num_ignored = 0
    
    #initial 10 percent points
    idx_10_percent = np.random.choice(range(x_train.shape[0]),size=int(0.1*x_train.shape[0]),replace=False)
    x_10 = x_train[idx_10_percent]
    y_10 = y_train[idx_10_percent]
    
    x_train = np.delete(x_train,idx_10_percent,0)
    y_train = np.delete(y_train,idx_10_percent,0)
    
    #initial_learning
    learner = ActiveLearner(estimator=SVC(probability=True),
                                X_training=x_10,y_training=y_10)
    while learner.score(x_tes, y_tes) < 0.97:
        #choose a point randomly to check - simulate a stream
        point = np.random.choice(range(len(x_train)))
        if classifier_margin(learner, x_train[point].reshape(1, -1)) <=  0.85:
            #learning again after every query; might avoid it if the trainnig is time consuming
            learner.teach(x_train[point].reshape(1, -1), y_train[point].reshape(-1, ))
            updated_acc = learner.score(x_tes, y_tes)
            accuracy_list.append(updated_acc)
            num_queried += 1
            print(f"Point updated: {point}; Accuracy achieved: {updated_acc}") 
        else:
            num_ignored += 1
            print(f"Ignored this point from labelling: {point}")
    return accuracy_list, num_queried, num_ignored
```

```python
def train_random_smapling(x_train,y_train,x_tes,y_tes):
    #to ensure uniformity of the stream
    
    np.random.seed(459)
    accuracy_list = []
    num_queried = 0
    num_ignored = 0
    
    #initial 10 percent points
    idx_10_percent = np.random.choice(range(x_train.shape[0]),size=int(0.1*x_train.shape[0]),replace=False)
    x_10 = x_train[idx_10_percent]
    y_10 = y_train[idx_10_percent]
    
    #initial_learning
    learner = ActiveLearner(estimator=SVC(probability=True),
                                X_training=x_10,y_training=y_10)
    while learner.score(x_tes, y_tes) < 0.97:

        #choose a point randomly to check - simulate a stream
        point = np.random.choice(range(len(x_train))) 
        learner.teach(x_train[point].reshape(1, -1), y_train[point].reshape(-1, ))
        updated_acc = learner.score(x_tes, y_tes)
        accuracy_list.append(updated_acc)
        num_queried += 1
        print(f"Point updated: {point}; Accuracy achieved: {updated_acc}")
        x_train = np.delete(x_train,point,0)
        y_train = np.delete(y_train,point,0)
    return accuracy_list, num_queried, num_ignored
```

```python
def train_entropy_sampling(x_train,y_train,x_tes,y_tes):
    np.random.seed(459)
    accuracy_list = []
    num_queried = 0
    num_ignored = 0
    #initial 10 percent points
    idx_10_percent = np.random.choice(range(x_train.shape[0]),size=int(0.1*x_train.shape[0]),replace=False)
    x_10 = x_train[idx_10_percent]
    y_10 = y_train[idx_10_percent]
    
    x_train = np.delete(x_train,idx_10_percent,0)
    y_train = np.delete(y_train,idx_10_percent,0)
    
    #initial_learning
    learner = ActiveLearner(estimator=SVC(probability=True),
                                X_training=x_10,y_training=y_10)
    
    print(np.mean(classifier_entropy(learner,x_train)))
    while learner.score(x_tes, y_tes) < 0.97:
        #choose a point randomly to check - simulate a stream
        point = np.random.choice(range(len(x_train)))
        if classifier_entropy(learner, x_train[point].reshape(1, -1)) >= 0.45:
            #learning again after every query; might avoid it if the trainnig is time consuming
            learner.teach(x_train[point].reshape(1, -1), y_train[point].reshape(-1, ))
            updated_acc = learner.score(x_tes, y_tes)
            accuracy_list.append(updated_acc)
            num_queried += 1
            print(f"Point updated: {point}; Accuracy achieved: {updated_acc}") 
        else:
            num_ignored += 1
            print(f"Ignored this point from labelling: {point}")
            
    return accuracy_list, num_queried, num_ignored
```

## Stream Based Sampling QBC

```python
def train_kl_divergence(x_train,y_train,x_tes,y_tes,n_committee):
    np.random.seed(459)
    
    accuracy_list = []
    num_queried = 0
    num_ignored = 0
    
    learners = []
    for i in range(n_committee):
        #initial 10 percent points
        idx_10_percent = np.random.choice(range(x_train.shape[0]),size=int(0.1*x_train.shape[0]),replace=False)
        x_10 = x_train[idx_10_percent]
        y_10 = y_train[idx_10_percent]
        
        learner = ActiveLearner(estimator=SVC(probability=True),X_training=x_10,y_training=y_10)
        learners.append(learner)
        
    committee = Committee(learner_list=learners)
    comm_perform = []
    comm_perform.append(committee.score(x_tes,y_tes))
    while committee.score(x_tes, y_tes) < 0.97:

        #choose a point randomly to check - simulate a stream
        point = np.random.choice(range(len(x_train)))
        if KL_max_disagreement(committee,x_train[point].reshape(1, -1)) >= 0.032:
            committee.teach(x_train[point].reshape(1, -1), y_train[point].reshape(-1, ))
            updated_acc = committee.score(x_tes, y_tes)
            comm_perform.append(updated_acc)
            num_queried += 1
            print(f"Point updated: {point}; Accuracy achieved: {updated_acc}")
        else:
            num_ignored += 1
            print(f"Ignored this point from labelling: {point}")
            
    return comm_perform, num_queried, num_ignored
```

```python
def train_vote_entropy(x_train,y_train,x_tes,y_tes,n_committee):
    np.random.seed(459)
    
    accuracy_list = []
    num_queried = 0
    num_ignored = 0
    
    learners = []
    for i in range(n_committee):
        idx_10_percent = np.random.choice(range(x_train.shape[0]),size=int(0.1*x_train.shape[0]),replace=False)
        x_10 = x_train[idx_10_percent]
        y_10 = y_train[idx_10_percent]
        
        learner = ActiveLearner(estimator=SVC(probability=True),X_training=x_10,y_training=y_10)
        learners.append(learner)
        
    committee = Committee(learner_list=learners)
    
    comm_perform = []
    comm_perform.append(committee.score(x_tes,y_tes))
    
    while committee.score(x_tes, y_tes) < 0.97:
        
        #choose a point randomly to check - simulate a stream
        point = np.random.choice(range(len(x_train)))
        if vote_entropy(committee,x_train[point].reshape(1, -1)) >= 0.027:
            committee.teach(x_train[point].reshape(1, -1), y_train[point].reshape(-1, ))
            updated_acc = committee.score(x_tes, y_tes)
            comm_perform.append(updated_acc)
            num_queried += 1
            print(f"Point updated: {point}; Accuracy achieved: {updated_acc}")
        else:
            num_ignored += 1
            print(f"Ignored this point from labelling: {point}")
            
    return comm_perform, num_queried, num_ignored
```

```python
def train_random_committee(x_train,y_train,x_tes,y_tes,n_committee):
    np.random.seed(459)
    
    accuracy_list = []
    num_queried = 0
    num_ignored = 0
    
    learners = []
    for i in range(n_committee):
        idx_10_percent = np.random.choice(range(x_train.shape[0]),size=int(0.1*x_train.shape[0]),replace=False)
        x_10 = x_train[idx_10_percent]
        y_10 = y_train[idx_10_percent]
        
        learner = ActiveLearner(estimator=SVC(probability=True),X_training=x_10,y_training=y_10)
        learners.append(learner)
        
    committee = Committee(learner_list=learners)
    comm_perform = []
    comm_perform.append(committee.score(x_tes,y_tes))

    while committee.score(x_tes, y_tes) < 0.97:
        #choose a point randomly to check - simulate a stream
        point = np.random.choice(range(len(x_train)))
        committee.teach(x_train[point].reshape(1, -1), y_train[point].reshape(-1, ))
        updated_acc = committee.score(x_tes, y_tes)
        comm_perform.append(updated_acc)
        num_queried += 1
        print(f"Point updated: {point}; Accuracy achieved: {updated_acc}")
        x_train = np.delete(x_train,point,0)
        y_train = np.delete(y_train,point,0)

    return comm_perform, num_queried, num_ignored
```
